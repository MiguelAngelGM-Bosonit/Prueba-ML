```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import pandas as pd
import numpy as np
```

```python
def load_and_select_data():
    data_df = pd.read_csv('./Empleados.csv')
    data_df = data_df[['DNI','Sexo','FechaNacimiento','Oficina','NivelIngles','CosteMesAnterior',
                       'tipo_incorporacion','parent_company','seniority','HorasSemanales','Situacion']]

    data_df = data_df.dropna()
    
    data_df = data_df.groupby('DNI', as_index = False).last().reset_index(drop = True)
    return(data_df)

Empleados = load_and_select_data()
```
