```python
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

seed = 14

import pandas as pd
import numpy as np

from sklearn import feature_extraction
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

```python
def load_and_select_data():
    data_df = pd.read_csv('./Empleados.csv')
    data_df = data_df[['DNI','Sexo','Edad','Oficina','NivelIngles','CosteMesAnterior',
                       'tipo_incorporacion','parent_company','seniority','HorasSemanales','Situacion']]

    data_df = data_df[~data_df.eq('unknown').any(1)]
    
    data_df = data_df.groupby('DNI').last().reset_index(drop = True)
    return(data_df)

Empleados = load_and_select_data()
```

```python
def one_hot(df):
    cat_columns = ['Sexo','Edad','Oficina','NivelIngles','tipo_incorporacion','parent_company','seniority']
    cat_dict = df[cat_columns].to_dict(orient='records')
    
    vec = feature_extraction.DictVectorizer()
    cat_vector = vec.fit_transform(cat_dict).toarray()

    df_vector = pd.DataFrame(cat_vector)
    vector_columns = vec.get_feature_names_out()
    df_vector.columns = vector_columns
    df_vector.index = df.index

    df_one_hot = df.drop(cat_columns, axis=1)
    df_one_hot = df_one_hot.join(df_vector)
    
    return(df_one_hot)

Empleados_one_hot = one_hot(Empleados)
Empleados_one_hot
```

```python
def label_to_binary_and_scale(df):  
    df['Situacion'] = df['Situacion'].map({'Baja Definitiva':0,'Activo':1})
    
    scaler = MinMaxScaler()
    scaled_df = scaler.fit_transform(df)

    scaled_df = pd.DataFrame(scaled_df,
                             columns = df.columns,
                             index = df.index)
    scaled_df = scaled_df.reset_index(drop = True)
    return(scaled_df)

Empleados_one_hot_scaled = label_to_binary_and_scale(Empleados_one_hot)
Empleados_one_hot_scaled
```

```python
def train_test(df):
    features = df.drop(['Situacion'],axis=1)
    labels = df['Situacion']

    X_train, X_test, y_train, y_test = train_test_split(features, labels,
                                                        test_size=0.20,
                                                        random_state=seed,
                                                        stratify=labels)
    return(X_train, X_test, y_train, y_test)

X_train, X_test, y_train, y_test = train_test(Empleados_one_hot_scaled)
```
